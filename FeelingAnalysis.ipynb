```python
from aip import AipNlp
import pandas as pd
import time
```

```python
AppID = 
ApiKey = 
SecretKey = 

aipnlp = AipNlp(AppID, ApiKey, SecretKey)
```

```python
data_path = "" # 读取csv文件
comments = pd.read_csv(data_path).iloc[:,0].tolist()
```

```python
# 人工智能组
a=0 # 积极评论数量
b=0 # 中性评论数量
c=0 # 消极评论数量
d=0 # 无效评论数量

for i in range(len(comments)):
    com_sen = comments[i]
    result = aipnlp.sentimentClassify(com_sen)
    if 'error_code' not in result:
        positive_prob = result['items'][0]['positive_prob']
        negative_prob = result['items'][0]['negative_prob']
        if positive_prob > negative_prob:
            a = a + 1
            print(comments[i] + "为积极评论。")
        elif positive_prob < negative_prob:
            c = c + 1
            print(comments[i] + "为消极评论。")
        else:
            b = b + 1
            print(comments[i] + "为中性评论。")
    else:
        # print("输入文本不符合情感分析类型")
        d = d + 1
        print("该评论为无效评论。")

# 输出结果
print("积极评论的数量为：",a)
print("消极评论的数量为：",c)
print("中性评论的数量为：",b)
print("无效评论的数量为：",d)
time.sleep(10)
```
