```python
import jieba
import wordcloud
import pandas as pd
import matplotlib.pyplot as plt
```

```python
def sen_read(file_path, output):
    sen = pd.read_csv(file_path).iloc[:,0]
    for i in range(len(sen)):
        sen_split = jieba.lcut(str(sen[i]))
        output.extend(sen_split)
    return output
```

```python
import collections
# 词频统计
def word_freq(input_list, output_list, stop_word):
    for word in input_list:
        if word not in stop_word:
            output_list.append(word)
    word_freq_result = collections.Counter(output_list).most_common(20) # 输出频率最高的前20个词
    return word_freq_result
```

```python
stop_word = {'，','？','！','\n','。','.',':','“','”',':',' ','不过','、','个','他们','：','真的','《','》','可以','很','上','时候','这么','来','一下','人','怎么','可能','只要','的话','但是','但','不是','打','没有','这个','还是','觉得','一个','他','了','才','并','更','者','着','的','我','是','就','不','也','都','被','说','你','在','笔记','置顶','那','记得','看打','这','把','年','给','真','还','有','和','唉','啊','哎','或者','吧','要','拿','看','再','最','跟','里','没','就是','到','吗','？','！','。','-'}
```

```python
"""
生成总词云图
"""
# 数据读取
result = []
for i in range(0,20,1):
    data_path = "WYY/WYY_rank"+str(i)+".csv"
    print("成功读取第"+str(i)+"个文件！")
    word = sen_read(file_path=data_path, output=result)

# 创建词云图字符串
wordcloud_input = " ".join(result)

wordcloud_result = wordcloud.WordCloud(
    font_path="C:/Windows/Fonts/simfang.ttf", # 设置字体
    stopwords=stop_word,
    background_color="white",
    max_words=200,
    width=1600,
    height=800,
).generate(wordcloud_input)
plt.imshow(wordcloud_result)
plt.savefig("WYY词云图.png")
plt.show()

# 整体词频统计
word_nostop = []
result = word_freq(input_list=result, output_list=word_nostop, stop_word=stop_word)
print(result)
```
