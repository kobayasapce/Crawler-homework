```python
from aip import AipNlp
import pandas as pd
import time
```

```python
def Feeling_Mul(id, a, b, c, d):
    data_path = "WYY/WYY_rank"+str(id)+".csv"
    print(data_path)
    comments = pd.read_csv(data_path).iloc[:,0].tolist()
    for i in range(len(comments)):
        com_sen = comments[i]
        result = aipnlp.sentimentClassify(com_sen)
        if 'error_code' not in result:
            positive_prob = result['items'][0]['positive_prob']
            negative_prob = result['items'][0]['negative_prob']
            if positive_prob > negative_prob:
                a = a + 1
            elif positive_prob < negative_prob:
                c = c + 1
            else:
                b = b + 1
        else:
            print("输入文本不符合情感分析类型")
            d = d + 1
    time.sleep(10)
    return a,b,c,d
```

```python
# 百度智能云自然语言处理模块参数
AppID = ''
ApiKey = ''
SecretKey = ''

aipnlp = AipNlp(AppID, ApiKey, SecretKey)
```

```python
# 人工智能组
a= 0 # 积极评论数量
b= 0 # 中性评论数量
c= 0 # 消极评论数量
d= 0 # 无效评论数量

for i in range(0, 20, 1):
    a1, b1, c1, d1  = Feeling_Mul(id=i, a=a, b=b, c=c, d=d)
    a = a1
    b = b1
    c = c1
    d = d1
    print("第"+str(i)+"轮运行完成！")
    print("积极评论的数量为：",a1)
    print("消极评论的数量为：",c1)
    print("中性评论的数量为：",b1)
    print("无效评论的数量为：",d1)
```
