```python
from selenium import webdriver
from selenium.webdriver.common.by import By
import time
import json
import pandas as pd
```

```python
# 登录并加载视频页面
def login_in(browser, url):
    # 进入要爬取的页面
    browser.get(url=url)
    f = open("bilibili_cookies.txt", 'r') # 保存着cookies的文件
    cookie = json.loads(f.read())
 
    # 利用cookie自动登录账号
    for i in cookie:
        browser.add_cookie(i)
    browser.refresh()
    time.sleep(10)
```

```python
def get_comment(url, save_path):
    number = 6
    browser = webdriver.Chrome()
    login_in(browser=browser, url=url)        
    # 加载当前页面出现了的所有评论
    comment_area = browser.find_element(By.CSS_SELECTOR, '#comment > div > div > div > div.reply-warp > div.reply-list')
    reply_items = comment_area.find_elements(By.CLASS_NAME, 'reply-item')
    comment_content_list = []
    like_list = []
    user_name_list = []
    for i in range(0,number):
        for item in reply_items:
            # 读取主评论的内容
            root_reply = item.find_element(By.CSS_SELECTOR, 'span[class="reply-content"]').text
            # if root_reply is None or i!=5:
            #     pass
            # else:
            #     comment_content_list.append(root_reply)
            # # 获取主评论的点赞数
            like = item.find_element(By.CSS_SELECTOR, 'span[class="reply-like"]').text
            # if like is None or i!=5:
            #     pass
            # else:
            #     like_list.append(like)
            if i ==number-1:
                comment_content_list.append(root_reply)
                like_list.append(like)
            print("主评论：" + root_reply)
            print(len(comment_content_list))
            print("====================")
        # 当前加载的评论全部获取完毕，需要滑动页面，并获得新加载的评论
        browser.execute_script("window.scrollTo(0, document.body.scrollHeight);")
        reply_items = comment_area.find_elements(By.CLASS_NAME, 'reply-item')
        print("第"+str(i)+"轮获取完成")
    output = pd.DataFrame({"content":comment_content_list, "like":like_list}).to_csv(save_path, sep=',', index=False)
    time.sleep(1)
    return 
```
